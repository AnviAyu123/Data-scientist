## Date Time Manipulation

```python
import numpy as np
import pandas as pd
```

```python
date_range = pd.date_range('4/12/2021',periods=12,freq='D')

dates = pd.DataFrame(date_range,columns=['Date'])

dates
```

```python
dates
```

```python
dates['Day'] = dates['Date'].dt.day
dates['Month'] = dates['Date'].dt.month
dates['Year'] = dates['Date'].dt.year

dates
```

#### Example to convert object date in to proper date format

```python
df = pd.DataFrame({'DATE_TIME':['Jan 1 12:00 am','Jan 2 12:00 am']})
df
```

```python
df['Date'] = pd.to_datetime(df['DATE_TIME'],format ='%b %d %I:%M %p')
df
```

#####  for different formats please visit     https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior

- or search for **_python datetime format documentation_**

```python
df['Day'] = df['Date'].dt.day
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year

df
```
