# Probabilty Problem

#### Problem - 1

- There are thousand student who's marks are captured based on their ability in the test .if the same test is given to a new student ,what is the probability of getting 50 marks.

```python
from numpy.random import randint as ri

import pandas as pd
```

```python
# Randomly generating 100 int b/w 30 to 100

marks = ri (30,101,1000)

stu_marks = pd.Series(marks)

stu_marks.head()
```

```python
# Let's take mean and standard deviation (std). We need these data to find the probability.

# Mean

marks_mean = stu_marks.mean()

print('mean marks = ',marks_mean)
```

```python
# Std 

marks_std = stu_marks.std()

print('marks std = ',marks_std)
```

```python
# What is the probability of a student scoring UPTO 50%

x = 50
```

## Probability always between 0-1 and starts from 0

```python
import scipy.stats as st 

# When X(data) is given then, use cdf function to find the probability of scoring marks up to 50(x)

prob = st.norm.cdf (x,loc = marks_mean ,scale = marks_std)

print(prob)
```

if we to find the probability of scoring marks greater than 50, we can minus prob (1-prob)

#### How it works

### Z distribution / statistics

Let's transform our normal distribution into Z distribution(Standard Normal Distribution). Where Mean = 0 and std =1

  **z= (x-mean)/std**

```python
score = (x-marks_mean) / marks_std

score
```
