###   Weak classifier - why tree?

**First what is a weak classifier?**

**_Weak classiffier_**  - slightly better than random guessing.

Any algorithm could have been used as a base for the boosting technique, but the reasion for choosing trees are:



**Pro's**

   - Computational scalability,
   - Handles missing values,
   - robuts to outliers,
   - does not require feature scaling,
   - can deal with irrelevant inputs,
   - interpretable (if small),
   - handles mixed predictors as well (quantitive and qualitative).
  
  
**Con's**

   - inability to extract a linear combination of features
   - high variance leading to a small computational power
   
and that's where boosting comes in to the picture. it minimises the variance by taking into consideration the results from various trees.
