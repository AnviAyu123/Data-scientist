```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
rupees_monthly = [2000,2530,900,15000,1500]
items = ['Tution_fee','Mobile_bill','Electricity_bill','Rent','Misc']


#plotting data

plt.pie(x= rupees_monthly, labels = items)
plt.show()
```

```python
plt.figure(figsize = (8,8))
plt.pie(x =rupees_monthly,labels = items,explode = [0.6,0,0.1,0,0],autopct = '%.2f', rotatelabels = True)
plt.show()
```
