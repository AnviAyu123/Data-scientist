# Bernoulli distribution

It is a type of discrete probability distribution.The Bernoulli Distribution essentially models a single trial of flipping a weighted coin. it is the probability distribution of a random variable taking on only two values, 1("success") and 0("Failure") with complementary probability p and 1-p respectively. the bernoulli distribution therefore describes events having exactly two outcomes, which are present in real life.

Suppose we have a single trial of with only two possible outcomes siccess or failure.

**P(x) = P^x * (1-p)^1 -x =p**

P(0) = P^0*(1-p)^1 = 1-p
